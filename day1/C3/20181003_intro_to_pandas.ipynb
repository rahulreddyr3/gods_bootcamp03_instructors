# 1. Importing Data

```
import pandas as pd
drinks = pd.read_csv('https://raw.githubusercontent.com/anthonytuanta/GA_DAT5/master/week2/lab/drink.csv',
                     na_filter=False, encoding='ISO-8859-1')
```

# 2. Basic Information of the data

```
# Basic info
drinks.info()
```

```
# Sneak peek of the data
drinks.head()
```

```
drinks.tail()
```

```
# For specific column
drinks['country'][0:10]
```

```
# Show more than 1 columns
drinks[['country','beer_servings']].head()
```

```

# I wanna see row x to y, column A to C

tmp_df = drinks.loc[10:20,'beer_servings':'wine_servings']
tmp_df
```

```
# Unique value 
drinks['country'].unique()
```

# 3. Basic Statistics of the data

By using describe() , we can know the basic statistics of the data for all numeric columns (count, mean, median, etc.)

```
drinks.describe()
```

# 4. Filtering with one or more conditions

To see the data in certain conditions


```
# By giving this condition, it will return a series of boolean 
[drinks.continent == 'NA']
```

```
# Using this series of boolean, it can filter the current data 
drinks[drinks.continent == 'NA'].head()
```

```
# For more than 1 conditions
drinks[(drinks.continent == 'NA') & 
       (drinks.wine_servings > 10) ].head()
```

# 5. Sorting Dataframe

```
drinks.sort_values(by = 'beer_servings', ascending = False)
```

```
# Combining filtering with sorting
drinks[drinks.continent == 'NA'].sort_values(by=['beer_servings', 'spirit_servings'],
                                             ascending=[False, True])
```

# 6. Using Aggregate Function

For example we would like to answer this question:
## What is the average wine consumption per person per year in Africa?

```


drinks[drinks.continent == 'AF']['wine_servings'].mean()
```

## Show a list of the top 10 spirit drinking countries.

```
spirits = drinks[['country', 'spirit_servings']]
spirits.sort_values(by='spirit_servings', ascending=False).head(10)
```

## Which continent has the highest on average wine consumption?

```
drinks[['wine_servings','continent']].groupby('continent').wine_servings.mean().sort_values(ascending=False).head(5)
```
